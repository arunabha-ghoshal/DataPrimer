```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

rng=np.random.RandomState(0)
x=np.linspace(0,10,500)
y=np.cumsum(rng.randn(500,6),0)

plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left')
```

```python
import seaborn as sns
sns.set()

plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left')
```
